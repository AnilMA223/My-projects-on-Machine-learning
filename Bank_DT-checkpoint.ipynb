Applying Machine Learning Algorithm Decision Tree.......

```python
df.info()
```

```python
#Saving in X and y
cols=['age','duration','campaign','pdays','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
X=df[cols]
y=df['y']
X.head()
```

```python
y[y=='no']=0
y[y=='yes']=1
y=pd.to_numeric(y)
y
```

```python
#Checking for datatypes again
```

```python
#Spliting the data......
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=46,test_size=0.3)
```

```python
y_train.dtypes
```

```python
#Importing necessary libraries
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
dc.fit(X_train,y_train)
predictions=dc.predict(X_test)
```

```python
#Checking for accuracy
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
